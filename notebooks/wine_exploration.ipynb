```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from src.data.load_wine import get_wine_data
```

```python
# Load the wine dataset
df_X, df_y = get_wine_data(as_frame=True)
df = df_X.copy()
df['target'] = df_y

df.head()
```

## Dataset Overview
Let's check the shape and basic statistics of the dataset.

```python
# Dataset shape and statistics
print(f"Dataset shape: {df.shape}")
df.describe()
```

## Class Distribution
Visualize the distribution of wine classes.

```python
# Plot class distribution with class names as labels
from sklearn.datasets import load_wine
class_names = load_wine().target_names
sns.countplot(x='target', data=df)
plt.title('Wine Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(ticks=range(len(class_names)), labels=class_names)
plt.show()
```

## Feature Correlation
Visualize the correlation matrix of the features.

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_X.corr(), annot=False, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

## Pairplot of Selected Features
Visualize pairwise relationships for a subset of features.

```python
# Pairplot for selected features
selected_features = ['alcohol', 'malic_acid', 'color_intensity', 'hue', 'target']
sns.pairplot(df[selected_features], hue='target', palette='Set1')
plt.show()
```
